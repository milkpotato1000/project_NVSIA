### **[통합]**

```python
class ArticleCrawling:

    def __init__(self, url):
        self.url = url
        self.soup = self._get_soup()

    def _get_soup(self):
        response = requests.get(self.url)
        return BeautifulSoup(response.text, 'html.parser')
    
    def get_article_url(self):
        meta_tag = self.soup.find("meta", property = "og:url")
        return meta_tag["content"] if meta_tag else None
    
    def get_id(self):
        url = self.get_article_url()    # https://www.spnews.co.kr/news/articleView.html?idxno=101049

        # source(spnew, ytn 등등)
        source = url.split("//")[1]
        source = source.split("/")[0]
        source = source.replace("www.", "").split(".")[0]

        # id
        if "idxno" not in url:
            return None
        idx = url.split("idxno=")[1]
        return f"{source}_{idx}"
    
    def get_title(self):
        header = self.soup.find("h1", class_ = "heading")
        return header.get_text(strip = True) if header else None
      
    def get_author(self):
        tag = self.soup.find("i", class_ = "icon-user-o")
        if tag:
            author = tag.parent.get_text(strip = True)
            return author.replace("기자명", "").strip()
        
    def get_category(self):
        breadcrumb = self.soup.find("ul", class_ = "breadcrumbs")
        if breadcrumb:
            for a in breadcrumb.find_all("a"):
                text = a.get_text(strip=True)
                if text in ["정치", "외교", "군사", "경제/산업", "사회/문화/체육"]:
                    return text
    
    def get_publish_date(self):
        tag = self.soup.find("i", class_ = "icon-clock-o")
        if tag:
            raw = tag.parent.get_text(strip = True)
            return raw.replace("입력", "").strip()
        
    def get_contents(self):
        article_body = self.soup.find_all("span", style = "font-size:18px;")
        contents = "\n".join(i.get_text(strip = True) for i in article_body)
        # return contents.rstrip("@") 
        return contents.split('@')[0].strip()
    
    def to_dict(self):
        return {
        # "ID": ,
        "url": self.get_article_url(),
        "id": self.get_id(),
        "title": self.get_title(),
        "author": self.get_author(),
        "category": self.get_category(),
        "publish_date": self.get_publish_date(),
        "contents": self.get_contents()
            # {
            #     "summary": ,
            #     "vector": ,
            #     "keywords": ,     >> 별개의 db로?
            # }
        }
    
    def __str__(self):
        data = self.to_dict()
        return "\n".join(f"{k}: {v}" for k, v in data.items())
```

### **[크롤링]**

```python
import time
import csv
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import pandas as pd

class SpnewsCrawler:
    def __init__(self,
                 start_idx: int,
                 end_idx: int,
                 output_csv: str = "spnews_articles.csv",
                 delay: float = 0.5):
        """
        start_idx: 시작 idxno (예: 101377)
        end_idx:   끝 idxno   (예: 100000 처럼 더 작은 값)
        output_csv: 저장할 CSV 파일 이름
        delay: 요청 사이 대기 시간(초)
        """
        self.base_url = "https://www.spnews.co.kr/news/articleView.html?idxno={}"
        self.start_idx = start_idx
        self.end_idx = end_idx
        self.output_csv = output_csv
        self.delay = delay

        # CSV가 없으면 헤더 생성
        try:
            open(self.output_csv, "r", encoding="utf-8-sig").close()
        except FileNotFoundError:
            with open(self.output_csv, "w", newline="", encoding="utf-8-sig") as f:
                writer = csv.writer(f)
                writer.writerow(["idxno", "url", "title", "date", "category", "content"])

    def get_article_url(self, idx: int) -> str:
        return self.base_url.format(idx)

    def fetch_html(self, url: str) -> str | None:
        """HTML 요청 (에러나 404 시 None 리턴)"""
        try:
            res = requests.get(url, timeout=10)
            # 404 같은 경우는 건너뛴다
            if res.status_code != 200:
                return None
            return res.text
        except requests.RequestException:
            return None

    def parse_article(self, html: str, url: str, idx: int) -> dict | None:
        """기사 HTML에서 필요한 정보 뽑기"""
        soup = BeautifulSoup(html, "html.parser")

        # --- 제목 ---
        # 실제 구조에 맞게 selector는 한 번 확인해서 바꿔줘야 할 수 있음
        # 예시 1: <h1 class="article-title">...</h1>
        title_tag = soup.select_one("h1")  # 필요시 'h1.article-title' 등으로 수정
        title = title_tag.get_text(strip=True) if title_tag else ""

        # --- 날짜 ---
        # 예시: <span class="date">2025-11-22 10:00</span>
        date_tag = soup.select_one(".date, .update, .article-date")
        date = date_tag.get_text(strip=True) if date_tag else ""

        # --- 카테고리(섹션) ---
        # 예시: <span class="category">정치</span>
        cat_tag = soup.select_one(".category, .position, .section")
        category = cat_tag.get_text(strip=True) if cat_tag else ""

        # --- 본문 내용 ---
        # 너가 쓰고 있던 것 반영: span style="font-size:18px;"
        article_body = soup.find_all("span", style="font-size:18px;")
        if article_body:
            contents = "\n".join(i.get_text(strip=True) for i in article_body)
        else:
            # 혹시 구조가 바뀐 경우 대비 예비 selector
            body_tag = soup.select_one(".article-body, .content, #articleBody")
            contents = body_tag.get_text("\n", strip=True) if body_tag else ""

        # 끝에 붙는 @ 제거
        contents = contents.rstrip("@")

        # 내용이 너무 없으면(삭제된 기사 등) 스킵 처리
        if not contents.strip():
            return None

        return {
            "idxno": idx,
            "url": url,
            "title": title,
            "date": date,
            "category": category,
            "content": contents
        }

    def append_to_csv(self, row: dict):
        with open(self.output_csv, "a", newline="", encoding="utf-8-sig") as f:
            writer = csv.writer(f)
            writer.writerow([
                row["idxno"],
                row["url"],
                row["title"],
                row["date"],
                row["category"],
                row["content"].replace("\r", " ").replace("\n", "\\n")  # 줄바꿈 처리
            ])

    def run(self, max_articles: int | None = None):
        """
        크롤링 실행
        max_articles: None이면 전체 범위, 숫자를 넣으면 그 개수까지만 수집
        """
        collected = 0

        # 이미 저장된 idxno 있으면 중복 방지용으로 불러오기 (선택사항)
        try:
            df_existing = pd.read_csv(self.output_csv, encoding="utf-8-sig")
            already = set(df_existing["idxno"].tolist())
        except Exception:
            already = set()

        # start_idx -> end_idx (내림차순) 순회
        step = -1 if self.start_idx > self.end_idx else 1
        idx_range = range(self.start_idx, self.end_idx + step, step)

        for idx in tqdm(idx_range, desc="크롤링 중"):
            if max_articles is not None and collected >= max_articles:
                print(f"\n지정한 {max_articles}개 수집 완료.")
                break

            if idx in already:
                continue

            url = self.get_article_url(idx)
            html = self.fetch_html(url)
            if html is None:
                time.sleep(self.delay)
                continue

            data = self.parse_article(html, url, idx)
            if data is None:
                time.sleep(self.delay)
                continue

            self.append_to_csv(data)
            collected += 1

            time.sleep(self.delay)

        print(f"\n총 {collected}개의 기사를 수집했습니다. → {self.output_csv}")
```

```python

if __name__ == "__main__":
    # ★ 여기만 네가 원하는 범위로 조정해서 돌리면 됨 ★

    # 예: 현재 마지막 기사 idxno가 101377이라고 가정하고
    # 우선 100개만 테스트 (101377 ~ 101278)
    START_IDX = 101377
    END_IDX = 101377 - 99

    crawler = SpnewsCrawler(
        start_idx=START_IDX,
        end_idx=END_IDX,
        output_csv="spnews_100_test.csv",
        delay=0.3
    )

    crawler.run(max_articles=100)
```
