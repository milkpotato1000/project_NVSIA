### 전처리 및 지오코딩

```python
import pandas as pd
import requests
from time import sleep

# 고유 지역을 다시 추출하고 전처리 규칙 적용
df['event_loc'] = df['event_loc'].fillna('')
all_raw_locations = []
for locations_str in df['event_loc']:
    # 쉼표로 분리하고 각 위치명에서 공백 제거
    locations = [loc.strip() for loc in locations_str.split(',') if loc.strip()]
    all_raw_locations.extend(locations)

processed_locations = []
for location in all_raw_locations:
    if location == '평안도':
        processed_locations.append('평안남도')
        processed_locations.append('평안북도')
    elif location == '황해도':
        processed_locations.append('황해북도')
        processed_locations.append('황해남도')
    elif location == '함경도':
        processed_locations.append('함경남도')
        processed_locations.append('함경북도')
    else:
        # '양강도'를 포함하는 경우 '량강도'로 변경
        if '양강도' in location:
            processed_locations.append(location.replace('양강도', '량강도'))
        else:
            processed_locations.append(location)

# 중복 제거 및 정렬하여 최종 고유 목록 생성
unique_locations = sorted(list(set(processed_locations)))

# '카타르'는 북한 지명이 아니므로 제외
unique_locations = [loc for loc in unique_locations if loc not in ['카타르']]

# (추가) 전처리 후 고유 지역 정보 출력
print(f"전처리 후 찾은 총 고유 지역 수: {len(unique_locations)}")
print("전처리 후 처음 10개 고유 지역:")
for i, loc in enumerate(unique_locations[:10]):
    print(f"{i+1}. {loc}")


# 지리 데이터 저장을 위한 딕셔너리 준비
location_geodata = {}

# 북한 지역의 검색 정확도를 높이기 위한 접두사
country_prefix = "조선민주주의인민공화국 "

# Nominatim API 엔드포인트
NOMINATIM_URL = "https://nominatim.openstreetmap.org/search"

# 고유 지역 목록을 반복하며 Nominatim API에 직접 쿼리
for i, location_name in enumerate(unique_locations):
    search_query = country_prefix + location_name
    print(f"위치 처리 중 {i+1}/{len(unique_locations)}: {search_query}")

    params = {
        'q': search_query,
        'format': 'geojson',
        'polygon_geojson': 1, # GeoJSON 폴리곤 데이터 요청
        'addressdetails': 1,
        'extratags': 1,
        'limit': 1 # 최적 일치 항목 하나만 필요
    }

    headers = {
        'User-Agent': 'my-geocoder-app-direct-polygon' # Nominatim에는 사용자 에이전트 필요
    }

    try:
        response = requests.get(NOMINATIM_URL, params=params, headers=headers)
        response.raise_for_status() # 4xx 또는 5xx 응답에 대한 HTTPError 발생
        data = response.json()

        if data and data['features']:
            # GeoJSON feature 객체에 geometry가 직접 포함됨
            geojson_data = data['features'][0]['geometry']
            location_geodata[location_name] = geojson_data
            print(f"  {location_name}에 대한 GeoJSON을 성공적으로 가져왔습니다.")
        else:
            print(f"  {location_name}에 대한 GeoJSON을 가져올 수 없습니다. (응답에 feature 없음 또는 위치를 찾을 수 없음)")
    except requests.exceptions.RequestException as e:
        print(f"  {location_name}에 대한 HTTP 요청 오류: {e}")
    except Exception as e:
        print(f"  {location_name}에 대한 응답 처리 오류: {e}")

    # API 사용 정책을 준수하기 위해 지연 시간 적용
    sleep(1) # 요청 사이에 1초 지연

print("\n지오코딩 완료.")
print(f"GeoJSON 데이터가 있는 총 지역 수: {len(location_geodata)}")

# 결과 샘플 표시 (처음 5개 항목)
print("\n추출된 GeoJSON 데이터 샘플:")
for i, (loc, geojson) in enumerate(location_geodata.items()):
    if i >= 5: # 간결성을 위해 처음 5개만 표시
        break
    print(f"위치: {loc}, GeoJSON 유형: {geojson.get('type')}, 좌표 샘플: {str(geojson.get('coordinates'))[:50]}...")

if not location_geodata:
    print("\n참고: GeoJSON 데이터가 검색되지 않았습니다. 이는 Nominatim에 특정 쿼리에 대한 폴리곤 데이터가 없거나 다른 API 제한 사항 때문일 수 있습니다.")
```

### 데이터프레임으로 저장(확인용)

```python
import pandas as pd

# location_geodata 딕셔너리에서 정보를 추출하여 리스트로 만듭니다.
data_for_df = []
for loc, geojson in location_geodata.items():
    geojson_type = geojson.get('type')
    coordinates_sample = str(geojson.get('coordinates'))[:100] + '...' if geojson.get('coordinates') else '좌표 없음'
    data_for_df.append({'위치': loc, 'GeoJSON 유형': geojson_type, '좌표 샘플': coordinates_sample})

# 추출된 정보를 Pandas DataFrame으로 변환합니다.
df_geodata = pd.DataFrame(data_for_df)

# DataFrame을 출력합니다.
display(df_geodata)
```

### 지도에 그려보기

```python
import folium
import random

# 한반도 중앙에 가까운 위도와 경도
latitude = 39.0
longitude = 127.0

# Folium 맵 객체 초기화
m = folium.Map(location=[latitude, longitude], zoom_start=7)

# 랜덤 색상 생성 함수
def get_random_color():
    return "#%06x" % random.randint(0, 0xFFFFFF)

# location_geodata 딕셔너리를 반복하며 GeoJSON 데이터 지도에 추가
for loc, geojson_data in location_geodata.items():
    # 각 지역에 대한 스타일 정의 (랜덤 색상)
    style_function = lambda x, color=get_random_color(): {
        'fillColor': color,
        'color': 'black',
        'weight': 1,
        'fillOpacity': 0.7
    }

    # folium.GeoJson 객체 생성 및 지도에 추가
    folium.GeoJson(
        geojson_data,
        name=loc,
        tooltip=folium.Tooltip(loc),
        style_function=style_function
    ).add_to(m)

# 지도를 HTML 파일로 저장하거나 직접 표시 (Jupyter Notebook/Colab 환경에서)
m
```
